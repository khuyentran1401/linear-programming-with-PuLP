```python
from pulp import *
# Create the 'prob' variable to contain the problem data
prob = LpProblem("Ambulance Allocation", LpMaximize)
```

```python
y = LpVariable.dicts("location", list(range(8)), cat='Binary')
```

```python
x = LpVariable.dicts("penaly", list(range(8)), cat='Binary')
```

```python
y[0]
```

```python
#Population of each location
pop = [40, 30, 35, 20, 15, 50, 45, 60]

#There must be at least one ambulance
b = [1,1,1,1,1,1,1,1]

#Define the number of ambulance
n = 1
```

Objective

```python
prob += lpSum(a[j,i]*pop[i]*y[j] for i in range(8) for j in range(8)) - 100000*lpSum(x[i] for i in range(8))
```

Constraint

```python
for i in range(8):
    prob += lpSum(a[i,j] * y[j] + a[i,j]*x[j] for j in range(8))  >= b[i]
```

```python
prob += lpSum(y[i] for i in range(8)) == n
```

```python
prob.solve()
```

```python
for i in range(8):
    print('location {}: {}, penalty {}: {}'.format(i+1,y[i].varValue, i+1, x[i].varValue))
```

```python
pulp.value(prob.objective) + 100000*sum(x[i].varValue for i in range(8))
```

```python
pulp.LpStatus[prob.status]
```

```python
prob.writeLP('prob.txt')
```

```python
y = LpVariable.dicts("location", list(range(8)), cat='Binary')
x = LpVariable.dicts("penalty", list(range(8)), cat='Binary')
```

```python
n1 = 2
```

```python
prob1 = LpProblem("Ambulance Allocation", LpMaximize)

for i in range (8):
    prob1 += lpSum(a[i,j] * y[j] + a[i,j]*x[j] for j in range(8))  >= b[i]

prob1 += lpSum(y[i] for i in range(8)) == n1

prob1 += lpSum(a[j,i]*pop[i]*y[j] for i in range(8) for j in range(8)) - 100000*lpSum(x[i] for i in range(8)) 

prob1.solve()

for i in range(8):
    print('location {}: {}, penalty {}: {}'.format(i+1,y[i].varValue, i+1, x[i].varValue))
```

```python
pulp.value(prob1.objective) +100000*sum(x[i].varValue for i in range(8))
```

```python
pulp.LpStatus[prob1.status]
```

```python
prob1.writeLP('prob1.txt')
```
